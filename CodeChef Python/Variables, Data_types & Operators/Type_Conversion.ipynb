```python
# ------------------------------------------------ Implicit Type Conversion -------------------------------------------------

# When Python encounters an operation involving different data types (like adding an integer and a float), it often performs
#   quick data type conversion. This is called Implicit type conversion or “type coercion.”

# Integer + Float:
num_int = 10
num_float = 3.5

result = num_int + num_float # Python automatically treats the integer 10 as a float 10.0 during addition.
print(f"{result}\n{type(result)}")   # Output: 13.5 & <class 'float'>

# ---------------------------------------------------------------------------------------------------------------------------------
# Boolean as Integer
bool_val = True   # In numeric context, True acts like 1
num_int  = 5

total = bool_val + num_int
print(f"{total}\n{type(total)}")  # Output: 6 & <class 'int'>
```

```python
# ------------------------------------------------ Explicit Type Conversion  -------------------------------------------------

# Explicit conversion requires manually conversion of data types using built-in functions. This is necessary when Python can’t
#       do it automatically or when you want more precise control.

# Common Type Conversion Functions:
    # --> int(): Converts data to an integer (e.g., "10" → 10 or 3.99 → 3).
    # --> float(): Converts data to a floating-point number (e.g., "3.14" → 3.14).
    # --> str(): Converts data to a string (e.g., 25 → "25").
    # --> bool(): Converts data to a boolean (e.g., 1 → True, 0 → False).

# ---------------------------------------------------------------------------------------------------------------------------------
# String to Integer:
num_str = "100"
num_int = int(num_str)

print(f"{num_int}\n{type(num_int)}")  # Output: 100 & <class 'int'>

# ---------------------------------------------------------------------------------------------------------------------------------
# Float to Integer:
num_float = 3.99
num_int = int(num_float)

print(f"{num_int}\n{type(num_int)}")   # Output: 3 & <class 'int'>

# ---------------------------------------------------------------------------------------------------------------------------------
# String to Float:
data_str = "45.67"
data_float = float(data_str)
print(f"{data_float}\n{type(data_float)}")   # Output: 45.67 & <class 'float'>

# ---------------------------------------------------------------------------------------------------------------------------------
# Integer to String:
age = 30
age_str = str(age)

print(f"{age_str}\n{type(age_str)}")  # Output: "30" & <class 'str'>

# ---------------------------------------------------------------------------------------------------------------------------------
# Boolean to Integer:
print(int(True))   # Output: 1
print(int(False))  # Output: 0
```
