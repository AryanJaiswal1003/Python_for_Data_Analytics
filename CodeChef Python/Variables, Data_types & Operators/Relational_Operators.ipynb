```python
""" ------------------------------------------------ Relational Operators -------------------------------------------------
## Contents:
    -- Equal To Operator (==) and Not Equal To Operator (!=)
    -- Greater Than Operator (>) and Less Than (<)
    -- Greater Than or Equal To (>=), Less Than or Equal To (<=)
    -- Comparing Multiple Values
"""
```

```python
# ------------------------------------------------ Equal To Operator (==) -------------------------------------------------

# The Equal To (==) and Not Equal To (!=) operators in Python belong to the relational operators family. They compare two
#       values and return a Boolean result: True or False.

# Equal To (==): Checks if two values match : Returns True if they’re the same, or False otherwise.
a = 5
b = 5
c = 10

print(a == b)  # Output: True (5 is equal to 5)
print(a == c)  # Output: False (5 is not equal to 10)

# ------------------------------------------------ Not Equal To Operator (!=) -------------------------------------------------

# Not Equal To (!=): Checks if two values are different : Returns True if they differ, or False if they’re the same.
x = 8
y = 12

print(x != y)  # Output: True (8 is not 12)
print(x != 8)  # Output: False (8 is the same as 8)
```

```python
# ------------------------------------------------ Greater Than (>) & Less Than (<) -------------------------------------------------

# Both are relational operators that compare two values, returning a Boolean result: True or False.

# A] Greater Than (>): Checks if the value on the left is greater than the value on the right: Returns True if the left value
#       exceeds the right value, otherwise False.
a = 10
b = 5
c = 10

print(a > b)  # Output: True  (10 is greater than 5)
print(b > a)  # Output: False (5 is not greater than 10)
print(a > c)  # Output: False (10 is not greater than 10, they're equal)

# ---------------------------------------------------------------------------------------------------------------------------------
# B] Less Than (<): Checks if the value on the left is less than the value on the right: Returns True if the left value is smaller,
#       otherwise False.
x = 3
y = 8
z = 3

print(x < y)  # Output: True  (3 is less than 8)
print(y < x)  # Output: False (8 is not less than 3)
print(x < z)  # Output: False (3 is equal to 3, not less)
```

```python
# ----------------------------------- Greater Than or Equal To (>=), Less Than or Equal To (<=) -----------------------------------

# A] Greater Than or Equal To (>=): Checks if the value on the left is greater than or equal to the value on the right: Returns
#       True if left ≥ right, otherwise False.
a = 10
b = 5
c = 10

print(a >= b)  # Output: True  (10 is greater than 5)
print(a >= c)  # Output: True  (10 equals 10)
print(b >= a)  # Output: False (5 is not ≥ 10)

# ---------------------------------------------------------------------------------------------------------------------------------
# B] Less Than or Equal To (<=): Checks if the value on the left is less than or equal to the value on the right: Returns True
#       if left ≤ right, otherwise False.
x = 3
y = 8
z = 3

print(x <= y)  # Output: True  (3 is less than 8)
print(x <= z)  # Output: True  (3 equals 3)
print(y <= x)  # Output: False (8 is not ≤ 3)

"""Can we use these operators with strings?
-- Ans: Yes, the "greater than or equal to" (>=) and "less than or equal to" (<=) operators can be used with strings in Python.
    When applied to strings, these operators perform a lexicographical comparison based on the Unicode values of the characters."""
```

```python
# ----------------------------------- Comparing Multiple Values -----------------------------------

# Relational operators can be chained together to compare multiple values in one concise expression. These chained
#       comparisons evaluate from left to right, allowing you to check several conditions at once and return a single Boolean
#           result: True or False.

# Chained Comparisons:
    # 1. a < b < c : Checks if a is less than b and b is less than c simultaneously.
    # 2. a > b > c : Checks if a is greater than b and b is greater than c simultaneously.
    # 3. a == b != c : Checks if a equals b and b does not equal c, all in a single statement.
a = 5
b = 3
c = 2

print(b < a > c) # Output: True  (3 < 5 and 5 > 2)
```
