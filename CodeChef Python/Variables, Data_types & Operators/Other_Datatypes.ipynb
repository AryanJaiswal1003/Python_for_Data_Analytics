```python
# ---------------------------------------------- Boolean Declaration ---------------------------------------------

# Booleans can be either True or False. They are treated like numbers, where True is equivalent to 1 and False is equivalent to 0.
#       This means you can use Booleans in math operations, such as adding them together.

# Declaring Boolean Variables:
is_active = True
is_logged_in = False

print(is_active)      # Output: True
print(is_logged_in)   # Output: False
print(type(is_active))  # Checking the data type --> Output: <class 'bool'>

# ---------------------------------------------------------------------------------------------------------------------------------
# Boolean in Numeric Context

print(True + True)    # Output: 2  (1 + 1)
print(False + True)   # Output: 1  (0 + 1)
print(False * 5)      # Output: 0  (0 * 5)

# Explanation:
    #1. True is equivalent to 1 in numeric operations.
    #2. False is equivalent to 0.
```

```python
# ---------------------------------------------- None Declaration ---------------------------------------------

# None is a special constant representing the absence of a value. It acts as a placeholder when you have no meaningful data to
#       assign to a variable or function parameter.

# Unlike False, 0, or an empty string, None belongs to its own distinct type, NoneType, and is not equal to any other object in Python.

# ---------------------------------------------------------------------------------------------------------------------------------
# A] Declaring a Variable as None:
x = None

print(x)        # Output: None
print(type(x))  # Output: <class 'NoneType'>

# ---------------------------------------------------------------------------------------------------------------------------------
# B] Common Use Cases

# 1. Placeholder Values: When you want to declare a variable but don't have a value for it yet.
user_name = None

# Later in the code:
user_name = "Alice"

# ---------------------------------------------------------------------------------------------------------------------------------
# 2. Clearing a Value: Assigning None can “reset” a variable, making it clear that no data is stored.
data = "Important info"

data = None
print(data)  # Output: None
```
