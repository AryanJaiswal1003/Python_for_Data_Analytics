```python
"""
------------------------------------------------- Logical Operators -------------------------------------------------
    ## Contents:
        -- AND Operator (and)
        -- OR Operator (or)
        -- NOT Operator

 Q] What is the Order of Precedence for logical operators?
 -- Ans:
    In Python, the order of precedence for logical operators is as follows (from highest to lowest):
        1. not
        2. and
        3. or
    So, if you have an expression like not a and b or c, it will be evaluated as ((not a) and b) or c.
"""
```

```python
# ------------------------------------------------- AND Operator (and) -------------------------------------------------

# The AND operator lets you combine multiple conditions into a single logical expression. It only evaluates to True if
#       all conditions are met. If even one condition is False, the entire expression returns False.

# Syntax: condition1 and condition2
    #1. Left-hand side (condition1) is evaluated first.
    #2. Only if the first condition is False does it skip checking the second condition (short-circuit evaluation).
    #3. The result is True only if both conditions are True; otherwise, it is False.

# ---------------------------------------------------------------------------------------------------------------------------------
"""What is short-circuit evaluation in AND?
-- Ans: Short-circuit evaluation in the context of the AND operator means that if the first condition is False,
    Python does not evaluate the second condition because the overall result will be False regardless of the second."""

number = 12
is_even = number % 2 == 0
is_large = number > 10

result = is_even and is_large
print(result)  # Output: True
```

```python
# ------------------------------------------------- OR Operator (or) -------------------------------------------------

# The OR operator evaluates multiple conditions and returns True if at least one of the conditions is True. If every
#       condition is False, it returns False.

# Syntax: condition1 or condition2
    # -- If either condition1 or condition2 is True, the entire expression becomes True.
    # -- If both are False, the result is False.
    # -- If the first condition is True, the second condition is NOT checked (short-circuit evaluation).

# ---------------------------------------------------------------------------------------------------------------------------------
number = 15
result = (number < 10) or (number > 20)

print(result)  # Output: False

"""What's the difference between OR and AND?
-- Ans: The OR operator returns True if at least one condition is True, while the AND operator returns True only
        if all conditions are True."""
```

```python
# ------------------------------------------------- NOT Operator (not) -------------------------------------------------

# The NOT operator inverts a Boolean value. If an expression is True, NOT turns it into False; if it’s False, NOT turns it into True.

flag = True
print(not flag)  # Output: False

# Explanation: flag holds True --> Applying not flips the value to False.
# ---------------------------------------------------------------------------------------------------------------------------------
x = 5
y = 10

condition = (x < y)  # True
print(not condition)  # Output: False3

# ---------------------------------------------------------------------------------------------------------------------------------
"""Check Password Strength Using not Operator: You have a password that must be considered “strong” if it meets two criteria:
    -- It must be longer than 8 characters.
    -- It must not have any spaces.
"""
password = "myp@ssword"  # User's password
is_strong = len(password) > 8 and password.count(' ') == 0 # Check if the password is strong using the 2 conditions

is_weak = not is_strong # Use the NOT condition to check if the password is weak
print(is_weak)
```
