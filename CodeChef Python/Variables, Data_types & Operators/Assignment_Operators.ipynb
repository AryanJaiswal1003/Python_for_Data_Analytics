```python
# ------------------------------------------------- Simple Assignment (=) -------------------------------------------------

# In Python, the = operator assigns a value to a variable. Think of a variable as a labeled box that can hold one piece of data
#       at a time, and the = sign as the instruction to place the data into that box.

x = 5  # x is now a variable that stores the integer value `5`.

# Reassigning Variables
x = 5
x = 10.5    # Changing from integer to float
x = "Hello" # Changing from number to string

# Note: Variables in Python are dynamic. You can store a different data type by simply assigning a new value.
```

```python
# ------------------------------------------------- Add and Assign (+=) -------------------------------------------------

# The operator += simplifies the process of updating a variable by adding a value to it and then assigning the result back to
#       the same variable.
x = 5
x += 3  # Equivalent to x = x + 3

print(x)  # Output: 8

# ------------------------------------------------- Substract and Assign (-=) -------------------------------------------------

# The operator -= simplifies the process of updating a variable by subtracting a value from it and then assigning the result
#       back to the same variable.
y = 10
y -= 4  # Equivalent to y = y - 4

print(y)  # Output: 6
```

```python
# ------------------------------------------------- Multiply and Assign (*=) -------------------------------------------------

# The operator *= simplifies the process of updating a variable by multiplying it by a value and then assigning the result
#       back to the same variable.
z = 3
z *= 2  # Equivalent to z = z * 2

print(z)  # Output: 6

# ------------------------------------------------- Divide and Assign (/=) -------------------------------------------------

# The operator /= simplifies the process of updating a variable by dividing it by a value and then assigning the result back to
#       the same variable.
a = 20
a /= 4  # Equivalent to a = a / 4

print(a)  # Output: 5.0
# Note: Itâ€™s important to note that you will always get a float result when using this operator, even if the division is exact.
```

```python
# ------------------------------------------------- Floor Divide and Assign (//=) -------------------------------------------------

# The floor division operator // performs integer (floor) division, meaning it divides two numbers and rounds down to the
#       nearest integer. The //= operator combines this floor division with assignment into a single, concise step.
x = 10
x //= 3

print(x)  # Output: 3
```
