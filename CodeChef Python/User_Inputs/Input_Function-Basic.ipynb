```python
# ------------------------------------------------ Basic input() Functions -------------------------------------------------

# The input() function in Python allows programs to take user input, making them interactive and dynamic.

# 1. Input as String: By default, input() captures user input as a string. The program waits for the user to enter a value
#       and press Enter, then stores it in a variable.
user_name = input("Enter your user_name: ")

print("Hello " + user_name + "!") # concatenation (`+`) connects text and variables(it should have string value)

# ---------------------------------------------------------------------------------------------------------------------------------
# 2. Input as Integer: Since input() always returns a string, numeric inputs must be converted to integers using int().
age = int(input("Enter a Number: "))

print("You are", age, "years old!") # we have age as int that's why we can not use concatenation (+)
```

```python
# ------------------------------------------------ Multiple Input on Same Line -------------------------------------------------

# Multiple inputs method is useful for collecting multiple pieces of information at once without prompting the user repeatedly.

# Using input() for Multiple Values --> The input() function waits for user input and stores it as a string. To accept
#       multiple values, we ask the user to separate them using a space or a comma.

# Example: Collecting First and Last Name:
user_input = input()  # User enters: John Doe
first_name, last_name = user_input.split()

print(f"First_name: {first_name}, Last_name: {last_name}")
# Here, .split() separates the input at spaces, storing "John" in first_name and "Doe" in last_name.

# ---------------------------------------------------------------------------------------------------------------------------------
"""How to handle inputs with different separators?
-- Ans: By specifying the separator in the .split() method. E.g., if Inputs are separated by commas, you would use input().split(',')."""
```

```python
# ------------------------------------------------ Multiple Input on Different Lines -------------------------------------------------
# Functions Involved --> .split() & map() + input()

#1. Using map() for Numeric Inputs: The map() function applies a conversion function (e.g., int()) to all elements in
#       an iterable. When combined with .split(), it processes multiple inputs efficiently in a single step.

numeric_inputs = map(int, input().split())  # Splits input and converts to integers

#This eliminates the need for multiple int() calls, ensuring clean and structured data handling.

# ---------------------------------------------------------------------------------------------------------------------------------
#2. Example of Structured Input:

string_inputs = input().split()      # input: Blue Mango --> User enters their favorite color and fruit on one line
numeric_inputs = map(int, input().split())      # input: 7 21 --> favorite number and lucky number on another line

# Assigning values
favorite_color, favorite_fruit = string_inputs[0], string_inputs[1]
favorite_number, lucky_number = numeric_inputs

print(f"Your favorite color is {favorite_color} and your favorite fruit is {favorite_fruit}.")
print(f"Your favorite number is {favorite_number} and your lucky number is {lucky_number}.")
```

```python
# Plan Your Weekend:

favorite_activity, favorite_place = input("Enter your Favorite activity & place: ").split()
num_friends, hours_spent = map(int, input("Enter num of friends & hours to spend: ").split())

# Convert activity and place to uppercase
favorite_activity = favorite_activity.upper()
favorite_place = favorite_place.upper()

print(f"You plan to go {favorite_activity} at the {favorite_place}.")
print(f"You will be joined by {num_friends} friends and will spend {hours_spent} hours there.")
```

```python
# ------------------------------------------------ Input Whitespace Handling -------------------------------------------------

# Whitespace in user input includes spaces, tabs, and newline characters that may be unintentionally entered by users.
#       Properly handling whitespace ensures that input is processed correctly.

#1. Handling Whitespace with .strip(); .rstrip() and .lstrip()
user_input = input("Enter Something with Spaces: ")
cleaned_input = user_input.strip()

print(f"Hello, '{cleaned_input}'!")
```
