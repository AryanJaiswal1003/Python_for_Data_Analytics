```python
# ------------------------------------------------ Filtering Invalid Characters -------------------------------------------------

# Using .strip() and .replace() for Cleaning Input -->
    #1. .strip() removes spaces from both ends of a string.
    #2. .replace() removes specific unwanted characters throughout the string.

# For example, with user input that has extra spaces and symbols:
user_input = input()  # Example: " @Hello! "

stripped_input = user_input.strip()  # Removes spaces
cleaned_input = stripped_input.replace("@", "").replace("!", "") # Cleaning the input

print(f"After .strip(): '{stripped_input}'")
print(f"After .replace(): '{cleaned_input}'")

# Here, .strip() removes extra spaces, and .replace() eliminates unwanted symbols (@, !).
```

```python
# ------------------------------------------------ Formatting User Input -------------------------------------------------

# Using f-string operator --> F-strings eliminate the need for concatenation or additional string methods, making it easier
#       to display structured and readable data.

#1. Formatting Numbers with f-Strings: When handling numeric input, f-strings allow precise decimal formatting using format specifiers.
price = float(input())  # Example input: 25.6789

print(f"The formatted price is: ${price:.2f}")

"""NOte:
The format specifier :.2f is used within an f-string to format a floating-point number (price in your example) to a specific
        number of decimal places.
    -- The : separates the variable name (price) from the format specification.
    -- .2 specifies that you want two digits after the decimal point.
    -- f indicates that the number should be displayed as a floating-point number."""


#2. Aligning and Padding Output Using f-Strings: For tables, reports, or structured output, f-strings help align text properly.
name = input()  # Example: John
score = int(input())  # Example: 96

print("\nFormatted Output:")
print(f"| {'Name':<10} | {'Score':>6} |")  # Header
print(f"| {name:<10} | {score:>6} |")  # User Data
```

```python
"""How to use different decimal precisions?
-- Ans: To use different decimal precisions, you simply change the number after the decimal point in the format
    specifier. For example:
        -- :.3f would format a number to three decimal places.
        -- :.4f would format a number to four decimal places."""

price = 12.34567

print(f"Price with two decimal places: {price:.2f}")
print(f"Price with three decimal places: {price:.3f}")
print(f"Price with four decimal places: {price:.4f}")
```
