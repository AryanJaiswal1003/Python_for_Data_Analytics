#### **➡️ NumPy String Functions Summary**

##### **➡️ Concatenation & Repetition**

| Function | Description |
|----------|-------------|
| `np.char.add()` | Adds two strings element-wise. |
| `np.char.multiply()` | Repeats each string element a given number of times. |

##### **➡️ Case Conversion**

| Function | Description |
|----------|-------------|
| `np.char.capitalize()` | Capitalizes the first letter of each string element. |
| `np.char.title()` | Converts each string element to title case. |
| `np.char.lower()` | Converts each string element to lowercase. |
| `np.char.upper()` | Converts each string element to uppercase. |
| `np.char.swapcase()` | Swaps the case of each string element. |

##### **➡️ Splitting & Joining**

| Function | Description |
|----------|-------------|
| `np.char.split()` | Splits each string element in the array (default by whitespace). |
| `np.char.splitlines()` | Splits each string element at line breaks. |
| `np.char.join()` | Joins elements in an array of strings using a specified separator. |

##### **➡️ String Searching**

| Function | Description |
|----------|-------------|
| `np.char.find()` | Returns the lowest index in each string where the substring is found (-1 if not found). |
| `np.char.rfind()` | Returns the highest index in each string where the substring is found (-1 if not found). |

##### **➡️ String Modification**

| Function | Description |
|----------|-------------|
| `np.char.strip()` | Removes leading and trailing whitespaces from each string element. |
| `np.char.lstrip()` | Removes leading whitespaces from each string element. |
| `np.char.rstrip()` | Removes trailing whitespaces from each string element. |
| `np.char.replace()` | Replaces occurrences of a substring in each string element. |

##### **➡️ Character Checking**

| Function | Description |
|----------|-------------|
| `np.char.isdigit()` | Checks if each string element contains only digits. |
| `np.char.isalpha()` | Checks if each string element contains only alphabetic characters. |
| `np.char.isnumeric()` | Checks if each string element contains only numeric characters. |
