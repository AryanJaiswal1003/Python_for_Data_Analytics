➡️ **Task: Combine Scores from Two Exams**

Write a program that:

1. Takes input of student scores from **two different exams**.  
2. Creates **two NumPy arrays** — one for each exam.  
3. Adds these arrays element-wise to get the **total scores** for each student.  
4. Displays the resulting array of total scores.

**`Input:`**
- **4 6 13 7 8**
- **3 15 9 1 3**

```python
import numpy as np

# Get input for exam scores
exam1_scores = list(map(int, input("Enter scores for Exam 1 (space-separated): ").split()))
exam2_scores = list(map(int, input("Enter scores for Exam 2 (space-separated): ").split()))

# Convert lists to numpy arrays
score1 = np.array(exam1_scores)
score2 = np.array(exam2_scores)

total_scores = score1 + score2 # Add the two arrays to get total scores
print(total_scores)
```

➡️ **Task: Recipe Ingredient Adjuster**

Write a Python program that:

1. Starts with a **NumPy array** representing the ingredient quantities for **4 servings**.  
2. Takes **user input** for the **desired number of servings**.  
3. Adjusts the ingredient quantities proportionally using **scalar division** or **multiplication**.  
4. Displays the adjusted ingredient quantities.

**`Input`: 6**

```python
import numpy as np

# Original recipe quantities for 4 servings
original_recipe = np.array([2, 3, 1, 4, 0.5])
desired_servings = int(input("Enter the desired number of servings: ")) # Get the desired number of servings from the user

scaling_factor = desired_servings / 4 # Calculate the scaling factor
adjusted_recipe = original_recipe * scaling_factor # Adjust the recipe quantities

print(adjusted_recipe)
```

➡️ **Task: Calculate Factorial Using NumPy Array Product**

Write a Python function that calculates the **factorial** of a given non-negative integer `n` using **NumPy’s product function**.

**`Input`: 5**

```python
import numpy as np

def calculate_factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        arr = np.arange(1, n+1) # Create a NumPy array of integers from 1 to n
        factorial = np.prod(arr) # Calculate the product of all elements in the array

    return factorial

n = int(input("Enter a non-negative number to calculate its factorial: "))
print(f"Factorial of {n} is {calculate_factorial(n)}")
```

➡️ **Calculate Cumulative Exam Scores**

You are tasked with calculating the cumulative exam scores for a student throughout a semester. Given a list of exam scores, create a NumPy array and **compute the cumulative sum of the scores**.

**`Input`: 67 72 79 84 89 63**

```python
import numpy as np

def calculate_cumulative_scores(scores):
    score = np.array(scores) # Convert the list to a NumPy array
    cumulative_scores = np.cumsum(score) # Calculate the cumulative sum of scores

    return cumulative_scores


scores = list(map(int, input("Enter exam scores separated by spaces: ").split())) # Get input from user

result = calculate_cumulative_scores(scores)
print(f"Cumulative exam scores: {result}")
```
