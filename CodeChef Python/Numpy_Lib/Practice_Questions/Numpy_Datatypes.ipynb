➡️ **Task: Even Number Boolean Mapper**

You are given a list of integers as input. Your task is to complete the NumPy program so that it returns a **boolean array** where:
- `True` represents **even numbers**, and  
- `False` represents **odd numbers**.  

**`Input`: 9 14 22 17 5**

```python
import numpy as np

def even_number_mapper(number):
    arr = np.array(number) # Convert list to Numpy array
    even_number = (arr % 2 == 0) # Check for boolean array where true indicates even number

    return even_number

numbers = input("Enter numbers seperated by spaces: ").split()
number_list = [int(num) for num in numbers]

result = even_number_mapper(number_list)
print(result)
```

➡️ **Task: Convert Numeric Scores to Letter Grades**

You are given a NumPy program that takes a list of **integer test scores**.  
Your task is to complete the function `convert_to_grades()` so that it converts the numeric test scores into **letter grades** based on the grading scale below:

| Score Range | Grade |
|--------------|--------|
| 90–100       | 'A'    |
| 80–89        | 'B'    |
| 70–79        | 'C'    |
| 60–69        | 'D'    |
| 0–59         | 'F'    |

**`Input`: 73 84 91 78 85 67**

```python
import numpy as np

def convert_to_grades(scores):
    # Create a NumPy array from the input list
    score_array = np.array(scores)

    # Create an empty string array of the same shape as score_array
    grade_array = np.empty_like(score_array, dtype='U1')

    # Your code here to fill grade_array with letter grades
    # grade_array[score_array >= 90] = 'A',
    # grade_array[(score_array >= 80) & (score_array < 90)] = 'B'
    # grade_array[(score_array >= 70) & (score_array < 80)] = 'C'
    # grade_array[(score_array >= 60) & (score_array < 70)] = 'D'
    # grade_array[score_array < 60] = 'F'

    grade_array = np.where(score_array >= 90, 'A',
                            np.where(score_array >= 80, 'B',
                            np.where(score_array >= 70, 'C',
                            np.where(score_array >= 60, 'D', 'F'))))

    return grade_array

# Get input from user
scores = input().split()
scores = [int(score) for score in scores]

# we can also use map to take input: scores = list(map(int, input().split()))

# Call the function and print the result
result = convert_to_grades(scores)
print(result)
```
