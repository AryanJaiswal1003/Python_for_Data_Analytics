➡️ **Task: Extract Every Third Element from an Array**

You are given a list of integers as input.  
Your task is to:

1. Convert the list into a **NumPy array**.  
2. Extract **every third element** from the array, starting from the first element (index `0`).  
3. Display the resulting array.

**`Input`: 4 6 -7 0 3 8**

```python
import numpy as np

# Get input from the user and convert it into numpy array of integers
input_list = input("Enter integers separated by spaces: ").split()
input_array = np.array(input_list, dtype=int)

# Extract every third element
result = input_array[::3]

print(result)
```

➡️ **Task: Reverse a NumPy Array Using Slicing**

You are given a NumPy array. Your task is to **reverse the order** of its elements using **array slicing**.

**`Input`: 9 14 20 7 3 5**

```python
import numpy as np

# Create a NumPy array from user input
input_list = input("Enter the integers separated by spaces: ").split()
input_list = [int(score) for score in input_list]
arr = np.array(input_list)

reversed_arr = arr[::-1] # Reverses the array using slicing
print(reversed_arr)
```

➡️ **Task: Mask Negative Numbers in a NumPy Array**

You are given a NumPy array containing both positive and negative numbers. Your task is to **create a boolean mask** to filter out negative numbers and display only the **non-negative** values.

**`Input`: 13 -11 8 7 -4 0 2**

```python
import numpy as np

# Get user input
input_string = list(map(int, input("Enter integers seperated by spaces: ").split()))
arr = np.array(input_string) # Create a NumPy array

# Create a boolean mask for non-negative numbers
boolean_mask = arr >= 0
result = arr[boolean_mask] # Apply the mask to get only non-negative numbers
print(result)
```

➡️ **Task: Filter Temperatures Above 50 and Even Numbers**

You are given a NumPy array representing temperatures in °F. Your task is to **filter and display only the even temperatures** that are **greater than 50°F** using Boolean indexing.

**`Input`: 34 42 50 47 57 54 60 21**

```python
import numpy as np

# Get temperature data from user input
temperatures = list(map(int, input("Enter temperatures in °F separated by spaces: ").split()))
temp_array = np.array(temperatures)

# Filter even temperatures above 50°F
filtered_temps = (temp_array > 50) & (temp_array % 2 == 0)
filtered_temps = temp_array[filtered_temps]

print(filtered_temps)
```

➡️ **Task: Create a Playlist from Song Indices**

You are given a NumPy array containing the names of 20 songs. Your task is to create a **playlist** by selecting specific songs using their **indices** provided by the user.

**`Input`: 3 6 8 13**

```python
import numpy as np

# Array of 20 songs
songs = np.array(['Song A', 'Song B', 'Song C', 'Song D', 'Song E',
                  'Song F', 'Song G', 'Song H', 'Song I', 'Song J',
                  'Song K', 'Song L', 'Song M', 'Song N', 'Song O',
                  'Song P', 'Song Q', 'Song R', 'Song S', 'Song T'])

# Get indices form user. # Create a playlist by selecting songs using their indices
indices = list(map(int, input("Enter song indices separated by spaces: ").split()))
songs = np.array([songs[index] for index in indices])
print(songs)
```
