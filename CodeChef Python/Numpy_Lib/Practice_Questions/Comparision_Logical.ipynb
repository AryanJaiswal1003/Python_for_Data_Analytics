➡️ **Task: Compare Athlete Performance with Qualifying Times**

**You are given two NumPy arrays:**
- `performance_times`: Represents the times recorded by athletes in a race.
- `qualifying_times`: Represents the minimum qualifying times required to advance.

**Your task is to:**
- Compare both arrays element-wise.
- Create and display a **boolean array** where `True` indicates the athlete **met or exceeded** the qualifying time, and `False` indicates they did not.

**`Input`: 13.3 23.2 45 11.2 49**

```python
import numpy as np

qualifying_times = np.array([10.2, 20.5, 45.0, 13.5, 49.5]) # Qualifying times

# Get athletes' performance times from user
athlete_times = np.array(list(map(float, input("Enter athletes' performance times separated by spaces: ").split())))
qualified = athlete_times <= qualifying_times # Compare athlete times with qualifying times

print(f"Athlete qualification status: {qualified}")
```

➡️ **Task: Combine Student Scores from Two Tests**

You are given two lists of student scores (ranging from 0 to 100) for two different tests. Your goal is to analyze which students performed well in **either test** or in **both tests**.

A score of **70 or above** is considered a **good performance**.

**`Input:`**
- **92, 85, 77, 61, 58**
- **81, 70, 64, 52, 88**

```python
import numpy as np

# Get comma-separated input for test scores
test1_scores = list(map(int, input("Enter test1 scores (comma-separated): ").split(',')))
test2_scores = list(map(int, input("Enter test2 scores (comma-separated): ").split(',')))

# Convert lists to NumPy arrays
arr1 = np.array(test1_scores)
arr2 = np.array(test2_scores)

# Create boolean masks for scores >= 70
mask1 = arr1 >= 70
mask2 = arr2 >= 70

# Use logical OR to find students who performed well in either test
performed_well_either = mask1 | mask2 # np.logical_or(mask1, mask2)
print(f"Performed well in either tests: {performed_well_either}")

# logical AND to find students who performed well in both tests
performed_well_both = mask1 & mask2 # np.logical_and(mask1, mask2)
print(f"Performed well in both tests: {performed_well_both}")
```

➡️ **Task: Compare Student Grades Across Subjects**

You need to create a program that compares student grades in **Math** and **Science** for multiple students.  
The goal is to identify:
- Students who performed well in **both** subjects.  
- Students who performed well in **at least one** subject.  

A score of **75 or above** is considered a **good performance**.

**`Input:`**
- **4**
- **67**
- **78**
- **83** 
- **89** 
- **91** 
- **87**
- **80** 
- **81** 

```python
import numpy as np

# Get the number of students
n = int(input("Enter the number of Students: "))

# Initialize arrays for Math and Science grades
math_grades = np.array([int(input("Enter Maths grade for student {}: ")) for i in range(n)])
science_grades = np.array([int(input("Enter Science grade for student {}: ")) for i in range(n)])

# Reshape the arrays to 2D (n x 1)
math_grades = math_grades.reshape(-1, 1)
science_grades = science_grades.reshape(-1, 1)

# Mask for students who scored 80 or above in both subjects
both_subjects_mask = np.logical_and((math_grades >= 80), (science_grades >= 80))
print(np.where(both_subjects_mask)[0] + 1) # Adding 1 to convert from 0-based to 1-based indexing

# Mask for students who scored 80 or above in at least one subject
either_subject_mask = np.logical_or((math_grades >= 80), (science_grades >= 80))
print(np.where(either_subject_mask)[0] + 1)  # Adding 1 to convert from 0-based to 1-based indexing
```

**Student Grade Filtering System**

Create a program that analyzes a 2D numpy array representing student grades in different subjects. The program should **identify students who have scored above a certain threshold in at least two subjects** and **below another threshold in at most one subject**

**`Input:`**
- **5**
- **3**
- **67 73 81**
- **83 87 91**
- **84 47 90**
- **48 73 84**
- **43 63 49**

```python
import numpy as np

# Input: Number of students and subjects
n_students = int(input("Enter the number of students: "))
n_subjects = int(input("Enter the number of subjects: "))

# Create a 2D numpy array for grades
grades = [list(map(int, input("Enter the grades of the students: ").split())) for i in range(n_students)]
grades = np.array(grades)

# Define thresholds
high_threshold = 80
low_threshold = 50

# Filter students based on the conditions
high_grade = grades > high_threshold
low_grade = grades < low_threshold

condition_1 = np.sum(high_grade, axis=1) >= 2 # Atleast 2 subjects with > 80
condition_2 = np.sum(low_grade, axis=1) <= 1 # Atmost one subject with < 80

# Gets indices of students meeting the criteria & [0] unpacks the tuple returned by np.where
result = np.where(condition_1 & condition_2)[0]
print(f"Students meeting the criteria: {result}")
```
