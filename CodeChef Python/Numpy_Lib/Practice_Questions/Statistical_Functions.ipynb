âž¡ï¸ **Task: Analyze Temperature Data with Mean and Median**

**Your task is to:**
1. Convert this list into a **NumPy array**.  
2. Calculate the **mean (average)** and **median** temperatures. 

**`Input`: 42.3 23 18.7 36.8 26.4**

```python
import numpy as np

def analyze_temperature(temperatures): # Function to calculate mean and median temperature
    temp_array = np.array(temperatures) # Converts the temp into an array

    mean_temp = np.mean(temp_array) # Calculate mean temperature
    median_temp = np.median(temp_array) # Calculate median temperature

    return mean_temp, median_temp

# Get daily temperature data from user
temperatures = list(map(float, input("Enter Daily Temperature Data (comma-separated): ").split()))

mean, median = analyze_temperature(temperatures)
print(
    f"Mean Temperature: {mean:.2f}\n"
    f"Median Temperature: {median:.2f}"
)
```

âž¡ï¸ **Task: Analyze Spread of Test Scores Across Subjects**

ðŸ”¹ You are given a **2D NumPy array** representing test scores of students in different subjects:  
- Each **row** â†’ a student  
- Each **column** â†’ a subject  

ðŸ”¹ Your task is to calculate the following **for each subject**:
- **Mean** (average score)  
- **Variance** (measure of spread)  
- **Standard Deviation** (how much scores deviate from the mean)

**`Input:`**
- **3**
- **5**
- **68 71 85 74 77**
- **74 85 75 82 82**
- **64 87 74 65 73**

```python
import numpy as np

def analyze_test_scores(scores):
    mean_scores = np.mean(scores, axis=0) # Calculate the mean scores for each subject
    variances = np.var(scores, axis=0) # Calculate the variance for each subject
    std_deviations = np.std(scores, axis=0) # Calculate the standard deviation for each subject

    return mean_scores, variances, std_deviations

# Get count of students and subjects from user
num_students = int(input("Enter the Number of Students: "))
num_subjects = int(input("Enter the Number of Subjects: "))

scores = [list(map(float, input("Enter the Students Scores across Subjects (space-separated): ").split())) for _ in range(num_students)]
scores_array = np.array(scores)

mean_scores, variances, std_deviations = analyze_test_scores(scores_array)

print(
    f"Mean Score: {mean_scores}\n"
    f"Variance: {variances}\n"
    f"Standard Deviation: {std_deviations}"
)
```

âž¡ï¸ **Task: Calculate Percentiles of Student Heights**

ðŸ”¹ **Your task is to:**
1. Convert the list into a **NumPy array**.  
2. Calculate the **25th, 50th (median), and 75th percentiles** of the heights.

ðŸ”¹Percentiles help understand the **distribution** of data â€” for example,  
- `25th percentile` â†’ height below which 25% of students fall  
- `50th percentile` â†’ median height  
- `75th percentile` â†’ height below which 75% of students fall  

**`Input`: 182 169 174 170 168**

```python
import numpy as np

def calculate_height_percentiles(heights):
    height = np.array(heights) # Convert the list to a NumPy array
    percentiles = np.percentile(height, [25, 50, 75]) # Calculate the 25th, 50th, and 75th percentiles

    return percentiles

# Get student heights input from user
heights = list(map(int, input("Enter the Student height (space-separated): ").split()))

result = calculate_height_percentiles(heights)
result = f"25th Percentile: {result[0]}\n50th Percentile: {result[1]}\n75th Percentile: {result[2]}"
print(result)
```

âž¡ï¸ **Calculate class performance metrics**

You are given a 2D NumPy array representing student scores in different subjects. Each **row represents a student**, and **each column represents a subject**. 

ðŸ”¹ **Your task is to calculate the following metrics:**
1. **Average score** for each subject
2. **Highest score** for each subject
3. **Overall pass percentage** (considering a score of 50 or above as passing)

```python
import numpy as np

def calculate_performance_metrics(scores):
    subject_averages = np.mean(scores, axis=0)
    subject_highest = np.max(scores, axis=0)

    passing_marks = scores >= 50
    overall_pass_percentage = np.mean(passing_marks) * 100

    return subject_averages, subject_highest, overall_pass_percentage

# Example usage
scores = np.array([
    [75, 80, 65, 90],
    [60, 70, 55, 85],
    [80, 75, 70, 95],
    [45, 60, 75, 70],
    [85, 90, 80, 88]
])

subject_averages, subject_highest, overall_pass_percentage = calculate_performance_metrics(scores)
print("Subject Averages:", subject_averages)
print("Subject Highest Scores:", subject_highest)
print("Overall Pass Percentage:", overall_pass_percentage)
```
